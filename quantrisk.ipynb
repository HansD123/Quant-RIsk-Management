**Extreme Value Theory:** <br>
Consider n random variables $X_i...X_n = S_n$, which represent financial losses.<br>
As accoridng to the Central Limit Theorem, for large n $S_n$ will converge into a normal distribution. <br>
This relies on $S_n$ being normalised through $(S_n - a_n/b_n)$ , where $a_n = nE(X_1)$ and $b_n = \sqrt{nvar(X_1)}$ <br>
The normalised maxima of the samples, $M_n = (X_1 .. X_n)$ fits the Generalised Extreme Value (GEV) distribution. <br>
GEV: $H_\xi(x) =
\begin{cases}
    \exp\left(-\frac{1}{\xi}(1 + \xi x)^{-1/\xi}\right) & \xi \neq 0, \\
    \exp(-e^{-x}) & \xi = 0.
\end{cases}$ <br><br>
$\xi$ is the shape parameter of the dsitribution. <br>$\xi$ = 0, Gumbel distribution.<br> $\xi <0$ , Weibull distribution. <br>$\xi > 0$ , Frechet distribution.

```python
#GEV distributiion:

def gev(x,xi):
    if xi == 0:
        return np.exp(-np.exp(-x))
    else:
        return np.exp(-(1/xi)*(1+xi*x)**(-1/xi))
    
xi_gumbel = 0
xi_weibull = -5
xi_frechet = 5

x_data = np.linspace(-5,5,1000)

figure,axes = plt.subplots(1,3,figsize = (15,5))

axes[0].scatter(x_data,gev(x_data,xi_gumbel))
axes[0].set_title("Gumbel Distribution")

axes[1].scatter(x_data,gev(x_data,xi_weibull),color = "orange")
axes[1].set_title("Weibull Distribution")

axes[2].scatter(x_data,gev(x_data,xi_frechet),color = "green")
axes[2].set_title("Frechet Distribution")

for axis in axes:
    axis.set_xlabel("x")
    axis.set_ylabel(r"$H_\xi (x)$")
    axis.grid()
plt.tight_layout()
plt.show()
```
